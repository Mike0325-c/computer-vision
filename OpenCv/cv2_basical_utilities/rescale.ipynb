```python
import cv2 as cv
import matplotlib.pyplot as plt
img = cv.imread('../Resources/Photos/cat.jpg')


print(img.shape)
def rescaleFrame(frame, scale=0.75):
    #image, videoes and live video
    width = int(frame.shape[1] * 0.3)
    height = int(frame.shape[0] * 0.8)
    
    
    # set the dimensions eauql to  tuple of width and height
    dimensions = (width,height)
    
    # resize a frame to the particular dimension
    # interpolation=cv.INTER_AREA is a parameter that specifies the interpolation method to be used when resizing an image
    # shrinking an image
    return cv.resize(frame,dimensions, interpolation =cv.INTER_AREA)

img_rescaled = rescaleFrame(img)
print(img_rescaled.shape)
# cv.imshow('cats_resize',img_rescaled)
# plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.xlabel('cat')
plt.imshow(img)

plt.subplot(1,2,2)
plt.imshow(img_rescaled)

plt.show()
# cv.waitKey(0)


```

```python
# reading video
capture = cv.VideoCapture("../Resources/Videos/dog.mp4")

while True:
    isTrue, frame = capture.read()
    
    frame_resized = rescaleFrame(frame)
    cv.imshow('Video',frame)
    cv.imshow('Video resized', frame_resized)
    # if cv.waitKey(20) & 0xFF==ord('d'):
    #     break
    
    if cv.waitKey(2) == 27:
        break
capture.release()
cv.destroyAllWindows()
```
