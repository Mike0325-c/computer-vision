```python
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread('../OpenCv/Resources/Photos/lady.jpg')

#rgb
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
# print(img)
print(img.shape)
plt.figure(figsize=(16,10))

plt.subplot(1,3,1)
plt.imshow(img_rgb )
plt.title('rgb_img')

#gray
img_gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
plt.subplot(1,3,2)
plt.imshow(img_gray)
plt.title('gray_img')


#otsu threshold

ret1, threshold1 = cv.threshold(img_gray, 0, 255, cv.THRESH_BINARY + cv.THRESH_OTSU)

# optimal threshold
print(ret1)
plt.subplot(1,3,3)
plt.imshow(threshold1)
plt.title('ostu_img')
plt.show()
```
